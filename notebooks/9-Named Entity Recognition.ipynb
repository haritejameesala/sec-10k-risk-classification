## Named Entity Recognition

```python
para = "Although most people consider piranhas to be quite dangerous, they are, for the most part, entirely harmless. Piranhas rarely feed on large animals; they eat smaller fish and aquatic plants. When confronted with humans, piranhas' first instinct is to flee, not attack. Their fear of humans makes sense, as far more piranhas are eaten by people than people are eaten by piranThe example above illustrates a well-structured paragraph, beginning with a topic sentence stating the harmless nature of piranhas. It follows with supporting details regarding their diet, natural instinct to flee from humans, and the fact that they are often prey for humans. The paragraph concludes by explaining that well-fed piranhas generally do not bite people."
```

```python
import nltk
words = nltk.word_tokenize(para)
```

```python
tags = nltk.pos_tag(words)
```
