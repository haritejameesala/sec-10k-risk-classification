```python
text = """
<html>
ITEM 1A. RISK FACTORS

Our business may be adversely affected by economic downturns.

ITEM 7. MANAGEMENT DISCUSSION AND ANALYSIS

Revenue increased by 10%.

...
"""
```

```python
import re

text1 = re.sub(r"<[^>]+>", " ", text)
print(text1)
```

```python
text2 = text1.lower()
print(text2)
```

```python
text3 = re.sub(r"[^a-zA-Z\s]", " ", text2)
print(text3)
```

```python
text4 = re.sub(r"\s+", " ", text3)
print(text4)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

words = text4.split()

text5 = " ".join(
    word for word in words
    if word not in stop_words
)
print(text5)
```

```python
def extract_risk_factors(text):

    start = text.lower().find("item 1a")
    end = text.lower().find("item 1b")

    if start != -1 and end != -1:
        return text[start:end]

    return text
```

```python
text6 = extract_risk_factors(text5)
print(text6)
```
