```python
_RE_HTML        = re.compile(r"<[^>]+>")
_RE_URL         = re.compile(r"https?://\S+|www\.\S+")
_RE_ITEM_HEADER = re.compile(r"\bitem\s+\d+[a-z]?\b\.?", re.IGNORECASE)
_RE_EXHIBIT_REF = re.compile(r"\bexhibit\s+\d+[\.\d]*\b", re.IGNORECASE)
_RE_BOILERPLATE = re.compile(
    r"(table of contents|annual report on form 10-k|"
    r"securities and exchange commission|"
    r"united states securities|"
    r"washington,?\s*d\.?c\.?|"
    r"incorporated by reference|"
    r"see note \d+|"
    r"f-\d+|"                          # financial statement page refs like F-12
    r"page \d+)",
    re.IGNORECASE
)
_RE_MONEY       = re.compile(r"\$[\d,\.]+\s*(million|billion|thousand)?", re.IGNORECASE)
_RE_PERCENT     = re.compile(r"\d+\.?\d*\s*%")
_RE_DATE        = re.compile(
    r"\b(january|february|march|april|may|june|july|august|"
    r"september|october|november|december)\s+\d{1,2},?\s+\d{4}\b",
    re.IGNORECASE
)
_RE_PURE_NUMBER = re.compile(r"\b\d[\d,\.]*\b")
_RE_WHITESPACE  = re.compile(r"\s+")
_RE_SPECIAL     = re.compile(r"[^a-z\s]")   # keep only lowercase letters + spaces
```

```python
def clean_text(text: str) -> str:
    """
    Full cleaning pipeline for a single section string.
    Order matters — run structural removals before character-level cleaning.
    """
    if not isinstance(text, str) or text.strip().lower() in ("nan", "none", ""):
        return ""

    t = text

    # 1. Remove HTML/XML tags (SEC EDGAR has embedded HTML)
    t = _RE_HTML.sub(" ", t)

    # 2. Remove URLs
    t = _RE_URL.sub(" ", t)

    # 3. Remove SEC boilerplate phrases
    t = _RE_BOILERPLATE.sub(" ", t)

    # 4. Remove ITEM headers (e.g. "ITEM 1A.", "Item 7.")
    t = _RE_ITEM_HEADER.sub(" ", t)

    # 5. Remove exhibit references
    t = _RE_EXHIBIT_REF.sub(" ", t)

    # 6. Remove dollar amounts  (noise for NLP; risk is in words not numbers)
    t = _RE_MONEY.sub(" ", t)

    # 7. Remove percentages
    t = _RE_PERCENT.sub(" ", t)

    # 8. Remove standalone dates
    t = _RE_DATE.sub(" ", t)

    # 9. Remove remaining pure numbers
    t = _RE_PURE_NUMBER.sub(" ", t)

    # 10. Lowercase
    t = t.lower()

    # 11. Remove all non-alphabetic characters (punctuation, symbols)
    t = _RE_SPECIAL.sub(" ", t)

    # 12. Collapse multiple spaces / newlines to single space
    t = _RE_WHITESPACE.sub(" ", t).strip()

    return t
```

```python
SECTION_COLS = {
    "risk":       "Risk Factors",
    "mda":        "Management’s Discussion and Analysis of Financial Condition and Results of Operations",
    "business":   "Business",
    "financials": "Financial Statements and Supplementary Data",
}
```

```python
def extract_sections(row: pd.Series) -> dict:
    """
    Pulls the 4 relevant columns from a DataFrame row.
    Returns a dict with raw text for each section.
    """
    return {
        "raw_risk":       str(row.get(SECTION_COLS["risk"],       "") or ""),
        "raw_mda":        str(row.get(SECTION_COLS["mda"],        "") or ""),
        "raw_business":   str(row.get(SECTION_COLS["business"],   "") or ""),
        "raw_financials": str(row.get(SECTION_COLS["financials"], "") or ""),
    }
```

```python
def build_combined_text(sections: dict) -> str:
    """
    Concatenates cleaned sections in priority order:
    Risk Factors > MD&A > Business > Financial Statements
    """
    cleaned = {k: clean_text(v) for k, v in sections.items()}

    parts = [
        cleaned["raw_risk"],
        cleaned["raw_mda"],
        cleaned["raw_business"],
        cleaned["raw_financials"],
    ]

    # Drop empty parts
    parts = [p for p in parts if p]
    return " ".join(parts)
```

```python
def assign_risk_score(text: str) -> float:
    # if not text or len(text.strip()) < 50:
    #     return 0.0

    text = text.lower()

    # ---------- Keyword Risk ----------
    total_words = max(len(text.split()), 1)

    high_count = sum(text.count(kw) for kw in HIGH_RISK_KEYWORDS)
    low_count  = sum(text.count(kw) for kw in LOW_RISK_KEYWORDS)

    high_density = (high_count / total_words) * 1000
    low_density  = (low_count / total_words) * 1000

    keyword_score = high_density - 0.5 * low_density

    # ---------- LM Sentiment Risk ----------
    tokens = lm.tokenize(text)
    lm_score = lm.get_score(tokens)

    polarity = lm_score["Polarity"]      # -1 to +1

    sentiment_score = -polarity * 10

    # ---------- Combined ----------
    risk_score = (
        0.6 * keyword_score +
        0.4 * sentiment_score
    )

    return risk_score
```

```python
df_raw["risk_score"] = label_text.apply(assign_risk_score)

q1 = df_raw["risk_score"].quantile(0.33)
q2 = df_raw["risk_score"].quantile(0.66)
```

```python
def assign_label2(score):
    if score <= q1:
        return "low_risk"
    elif score <= q2:
        return "medium_risk"
    else:
        return "high_risk"
```
