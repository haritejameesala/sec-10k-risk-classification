## Wordnet Lemmatizer

### like stemming
### Output: "Lemma"(Root Word) XXX not Root Stem(output of stemming)

```python
from nltk.stem import WordNetLemmatizer
# more time than stemming
lemmatizer = WordNetLemmatizer()
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
"""
Noun - n
Verb -v
adjective - a
Adverb - r
"""
lemmatizer.lemmatize("going" ,pos = 'v')
```

```python
words = ["finally", "finalized","goes", "going", "go","gone","congratulations","sportingly","fairly","history"]
```

```python
for word in words:
    print(word+ "--->"+ lemmatizer.lemmatize(word))
```
